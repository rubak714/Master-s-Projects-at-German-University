## Combination of datasets containing 10 genres only

```python
# Stacking the dataframes on top of each other

genre16_dataset=pd.concat([acoustic_df,blues_df,disco_df,guitar_df,romance_df,kpop_df,comedy_df,chill_df,anime_df,afrobeat_df,british_df,country_df,dance_df,hardcore_df,malay_df,worldmusic_df], axis=0)
```

## Shuffling in pandas dataframe by resetting index

```python
df=genre16_dataset.sample(frac=1).reset_index(drop=True)
```
